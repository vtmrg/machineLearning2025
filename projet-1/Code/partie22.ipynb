```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Carregar os dados
data = pd.read_excel("Sujet1.xlsx", sheet_name=0, header=0, index_col=0)

# Separar features e labels
X_features = data[['delta', 'theta', 'alpha', 'sigma', 'beta', 'puissance']]
y_labels = data['phase']

X_apprentissage = X_features.iloc[0:420]
y_apprentissage = y_labels.iloc[0:420]

# validação
X_validation = X_features.iloc[420:]
y_validation = y_labels.iloc[420:]

```

```python
# Contagem das classes em treino e validação
counts_train = y_apprentissage.value_counts().sort_index()
counts_val = y_validation.value_counts().sort_index()

# Criação do DataFrame para comparação
df_counts = pd.DataFrame({
    'X_apprentissage': counts_train,
    'X_validation': counts_val
}).fillna(0)  # preencher com 0 caso alguma fase não exista em algum conjunto

# Plotagem do histograma
df_counts.plot(kind='bar', figsize=(10,6))
plt.title("Répartition des phases de sommeil - X_apprentissage vs X_validation")
plt.xlabel("Phase de sommeil")
plt.ylabel("Nombre d'occurrences")
plt.xticks(rotation=0)
plt.legend(title="Ensembles")
plt.tight_layout()
plt.show()
```

```python
# Contagem das classes em X_apprentissage
counts_train = y_apprentissage.value_counts()

# Plotando o gráfico de pizza
plt.figure(figsize=(8,8))
plt.pie(counts_train, labels=counts_train.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.tab20.colors)
plt.title("Proporção das fases de sono em X_apprentissage")
plt.axis('equal')  # para deixar o círculo perfeito
plt.show()

# Contagem das classes em X_apprentissage
counts_train = y_validation.value_counts()

# Plotando o gráfico de pizza
plt.figure(figsize=(8,8))
plt.pie(counts_train, labels=counts_train.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.tab20.colors)
plt.title("Proporção das fases de sono em X_apprentissage")
plt.axis('equal')  # para deixar o círculo perfeito
plt.show()
```
